```python
# Import modules
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

```python
# Sample data for email spam classification
y_true = np.array([1] * 50 + [0] * 50)  # 50 spam (1) and 50 not spam (0)
y_pred = np.array([1] * 40 + [0] * 10 + [1] * 5 + [0] * 45)  # Simulated Model's predictions

# Generate confusion matrix
cm = confusion_matrix(y_true, y_pred)
print("Confusion Matrix", "\n", cm)
```

```python
# Create figure and axes for side-by-side heatmaps
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Heatmap with labels
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Not Spam", "Spam"], yticklabels=["Not Spam", "Spam"], ax=axes[0])
axes[0].set_title("Confusion Matrix with Labels")
axes[0].set_xlabel("Predicted Label")
axes[0].set_ylabel("Actual Label")

# Heatmap without labels
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", ax=axes[1])
axes[1].set_title("Confusion Matrix without Labels")
axes[1].set_xlabel("Predicted Label")
axes[1].set_ylabel("Actual Label")

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Calculate metrics
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)

# Print results
print(f"Accuracy: {accuracy:.2%}")
print(f"Precision: {precision:.2%}")
print(f"Recall: {recall:.2%}")
print(f"F1 Score: {f1:.2%}")
```

```python
from sklearn.metrics import classification_report

report = classification_report(y_true, y_pred)
print(report)
```
