Linear modeling is based on the assumption that the relationship between the dependent variable \( y \) and the independent variable(s) \( x \) is linear and can be represented as:

$y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_n x_n + \epsilon$

where:

- **( y \)** = dependent (target) variable
- **( x_1, x_2, ..., x_n \)** = independent (predictor) variables
- **( $\beta_0$ \)** = intercept (value of \( y \) when all \( x \) are 0)
- **( $\beta_1, \beta_2, ..., \beta_n$ \)** = regression coefficients (weights for predictors)
- **( $\epsilon$ \)** = error term (captures randomness and unobserved factors)

### Types of Linear Models

#### **1. Simple Linear Regression**
A linear model with one independent variable.

$y = \beta_0 + \beta_1 x + \epsilon$

#### **2. Multiple Linear Regression**
A linear model with more than one independent variable.

$y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_n x_n + \epsilon$

### Ordinary Least Squares (OLS)
OLS is a method for estimating the parameters (\( $\beta_0, \beta_1, ..., \beta_n$ \)) of a linear model by minimizing the sum of squared residuals:

$\sum_{i=1}^{n} (y_i - \hat{y}_i)^2$

where:

- $y_i$ = actual observed value
- $\hat{y}_i$ = predicted value from the model

OLS finds the best-fit line (or hyperplane) by minimizing this error.

### Assumptions of OLS
For OLS to provide unbiased and efficient estimates, the following assumptions must hold:

1. **Linearity**: The relationship between \( x \) and \( y \) is linear.
2. **Independence**: Observations are independent of each other.
3. **Homoscedasticity**: The variance of residuals is constant across all levels of \( x \).
4. **Normality**: The residuals follow a normal distribution.
5. **No Multicollinearity**: Independent variables are not too highly correlated.

### Advantages:
- Simple and interpretable.
- Efficient under the right conditions.
- Works well when assumptions hold.

### Limitations:
- Sensitive to outliers.
- Assumes a linear relationship (real-world data is often nonlinear).
- Poor performance with multicollinearity or heteroscedasticity.

### Summary
- **Linear modeling** describes the relationship between dependent and independent variables using a straight line (or hyperplane).
- **OLS** is a method for estimating the parameters of a linear model by minimizing squared residuals.
- **Assumptions** must hold for OLS to produce unbiased and efficient estimates.
- **Python implementations** are available in `statsmodels` and `scikit-learn`.

## Linear vs Cubic Relationship:

In the real world, the relationship between 2 things is almost always nonlinear i.e. the Rate of Change or slope is not a universal constant. Then why do we invest time in Simple linear regression? Because it is highly interpretable though it has low flexibility, while cubic regression is more flexible but less interpretable. Since cubic regression can capture more complex patterns in the data, its training RSS will always be lower or equal to that of linear regression. However, this increased flexibility can lead to overfitting, as the model may also capture noise rather than just the true underlying trend.

## Regression Dilution:

Regression dilution happens when measurement error in X weakens the estimated relationship between X and Y, for example students incorrectly reporting their hours of studies for exams. This causes the slope estimate to be biased, making the effect seem different than it really is. It leads to inaccurate conclusions and weaker predictions. Using more precise measurements or statistical corrections can help reduce this bias.

When measurement errors are present in either predictor or response variables, traditional OLS regression may not be appropriate. Models like Total Least Squares, Deming Regression, and Weighted Least Squares are designed to account for these uncertainties, leading to more reliable parameter estimates.

## Ordinary Least Square (OLS) and Maximum Likelihood Estimation (MLE):

In the context of linear regression models, Least Squares (OLS) and Maximum Likelihood Estimation (MLE) both lead to the same parameter estimates under normally distributed errors, but they use different approaches.
  
 - Least Squares focuses on minimizing the sum of squared differences between observed and predicted values, finding the best-fitting line.
 - MLE focuses on maximizing the likelihood of the observed data, estimating parameters that make the given data most probable.
  
In summary, Least Squares is a geometric approach to fitting a line, while MLE is a statistical approach based on probability maximization.

```python
#IMPORT LIBRARIES
from pprint import pprint
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sklearn.model_selection import train_test_split
from scipy.optimize import fsolve
```

```python
#READ AUTO.CSV DATASET
df_auto = pd.read_csv('Auto.csv')

pprint(df_auto.head(5))
pprint(df_auto.info())
```

```python
#PERFORMING PRE PROCESSING OF DATA.

print(f"\n{df_auto.info()}")
# We need to change the dtatype of horsepower to int which
print(f"\n{df_auto.describe(include='all')}")
# There are no null values
print(f"\n{df_auto.isnull().sum()}")

# We need to change the type of hosepower from object to integer.
print(f"\n{df_auto['horsepower'].unique()}")

# There are some values containing question marks in horsepower, we need to fix them.

# 2. Identify rows where horsepower == '?'
missing_hp_rows = df_auto[df_auto['horsepower'] == '?']
print(f"\nRows with missing horsepower:")
print(missing_hp_rows)

# 3. Replace '?' with NaN
df_auto.loc[df_auto['horsepower'] == '?', 'horsepower'] = np.nan

# 4. Convert horsepower column to float
df_auto['horsepower'] = df_auto['horsepower'].astype('float')

# 5. Fill missing horsepower with the mean HP for that car name
df_auto['horsepower'] = df_auto.groupby('name')['horsepower'].transform(
    lambda x: x.fillna(x.mean())
)

# For the remaining values We might want to fill those remaining NaNs with a global mean.
df_auto['horsepower'] = df_auto['horsepower'].fillna(
    df_auto['horsepower'].mean()
)

# 4. Convert horsepower column to float
df_auto['horsepower'] = df_auto['horsepower'].astype('int64')

print(f"\nAfter handling NAN or ? values:")
print(df_auto.isnull().sum())
```

```python
df_auto.describe(include='all').T
```

```python
columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'year', 'origin']

# Set figure size
plt.figure(figsize=(30, 20))

# Create subplots for each column
for i, col in enumerate(columns, 1):
    plt.subplot(3, 3, i)  # Creating a 3x3 grid (adjust if needed)
    sns.boxplot(y=df_auto[col])
    plt.title(col, fontsize=24)

plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

```python
# Define predictor (X) and response (y)
X = df_auto[['horsepower']]
y = df_auto['mpg']

# We will not use Train Test Split here
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=4)

# Add constant term for intercept
#X_train = sm.add_constant(X_train)
#X_test = sm.add_constant(X_test)
X = sm.add_constant(X)

# Fit the regression model on training data
#model = sm.OLS(y_train, X_train)
model = sm.OLS(y, X)

results = model.fit()

# Print the summary of regression results
print(results.summary())
```

3. The negative slope (-0.1575) tells us that as horsepower increases, fuel efficiency (mpg) decreases. Specifically, for every 1-unit increase in horsepower, mpg drops by 0.1575. This confirms that cars with more powerful engines tend to use more fuel per mile.

### Key Takeaway:

The model is statistically significant and the relationship between horsepower and mpg is moderately strong. Additional predictors like weight, cylinders, acceleration can further improve the model.

### Now we will Use the `ols()`  function from statsmodels library to perform a multiple linear regression with mpg as the response and all other variables except name as the predictors. 

```python
# Select predictor variables (exclude 'mpg' and 'name' column)
X = df_auto.drop(columns=['mpg', 'name'])

# Convert categorical variables (if any) to numeric (e.g., one-hot encoding for 'origin' if needed)
if 'origin' in X.columns:
    X = pd.get_dummies(X, columns=['origin'], drop_first=True)  # Convert 'origin' to dummy variables

# Define the response variable
y = df_auto['mpg']

X['origin_2'] = X['origin_2'].astype(int)
X['origin_3'] = X['origin_3'].astype(int)
X = X.apply(pd.to_numeric)  # Ensures all values are numeric

# Add a constant term for the intercept
X = sm.add_constant(X)

# Fit the multiple linear regression model
model = sm.OLS(y, X).fit()

# Print the summary of the regression model
print(model.summary())
```

There is a significant relationship between predictors and response.

- Displacement, Weight, Year, and Origin significantly impact the response.
- Cylinders Horsepower and Acceleration do not significantly contribute.
- The F-statistic is 228.8 which is > 1. 
- The probability of F-statistics is < 0.05.
- This confirms that there is a strong relationship between one or many of the predictors variables and response variable. 

The p-values associated with each predictors t-statistic indicate that "displacement", "weight", "year", and "origin" have a statistically significant relationship with mpg, while "cylinders", "horsepower", and "acceleration" do not.

The coefficient of the "year" variable suggests that, on average, for each additional year, mpg increases by 0.7894, assuming all other predictors remain constant.

### Now we will Use the `plot_fit ()` function from statsmodels to produce diagnostic plots of the linear regression fit.

```python
# Number of predictors (excluding intercept)
num_predictors = len(model.params) - 1  

# Ensure grid size adjusts dynamically (Max 3 columns)
rows = (num_predictors // 3) + (num_predictors % 3 > 0)
cols = min(3, num_predictors)

fig, ax = plt.subplots(rows, cols, figsize=(12, 6))

count = 1  # Start from first predictor

for i in range(rows):
    for j in range(cols):
        if count <= num_predictors:
            sm.graphics.plot_fit(model, count, ax=ax[i, j])

            # Set subplot title for better readability
            ax[i, j].set_title(f"Fitted Plot for {model.model.exog_names[count]}", fontsize=10)
            count += 1
        else:
            ax[i, j].axis("off")  # Hide unused subplots

# Adjust layout
fig.tight_layout()
plt.show()
```

```python
# Create a 2x2 grid for diagnostic plots
fig, axes = plt.subplots(2, 2, figsize=(15, 15))

# Residuals vs. Fitted
axes[0, 0].scatter(model.fittedvalues, model.resid, alpha=0.5)
axes[0, 0].axhline(y=0, color='red', linestyle='--')  # Horizontal reference line
axes[0, 0].set_xlabel("Fitted values")
axes[0, 0].set_ylabel("Residuals")
axes[0, 0].set_title("Residuals vs. Fitted")

# Histogram of Residuals
sns.histplot(model.resid, bins=20, kde=True, ax=axes[0, 1])
axes[0, 1].set_xlabel("Residuals")
axes[0, 1].set_ylabel("Frequency")
axes[0, 1].set_title("Histogram of Residuals")

# Q-Q Plot (Check for normality)
sm.qqplot(model.resid, line='s', ax=axes[1, 0])
axes[1, 0].set_title("Q-Q Plot")

# Leverage vs. Residuals Plot
sm.graphics.plot_leverage_resid2(model, ax=axes[1, 1])  # High leverage detection
axes[1, 1].set_title("Leverage vs. Residuals Plot")

# Adjust layout
plt.tight_layout()
plt.show()
```

## Now using this data, Lets fit a least squares regression to predict $y$ using $x_1$ and $x_2$. And answer, what are $_0$, $_1$, and $_2$? How do these relate to the true $_0$, $_1$, and $_2$? Can we reject the null hypothesis $H_0$: $_1$ = 0? How about the null hypothesis $H_0$: $_2$ = 0?


```python
df_x1 = pd.DataFrame({'x1': x1})
df_x2 = pd.DataFrame({'x2': x2})
df_y = pd.DataFrame({'y': y})

df_x_q7c = pd.concat([df_x1, df_x2], axis=1)
df_x_q7c = sm.add_constant(df_x_q7c)

# Fit the regression model
model7c = sm.OLS(df_y, df_x_q7c).fit()

# Print the model summary
print(model7c.summary())
```

## Model Analysis and Multicollinearity Impact

The estimated coefficients are $B_0$ $\approx$ 2.18, $B_1$ $\approx$ 0.70, and $B_2$ $\approx$ 2.50. While the intercept is close to the true value 2, the coefficients for $X_1$ and $X_2$ differ significantly from their expected values.  

- The high p-value for $B_1$ suggests that y does not significantly depend on $x_1$.  
- The low p-value for $B_2$ confirms that y depends on $x_2$.  
- Although $x_1$ and $x_2$ are correlated, their expected impact on y is not well captured in the model.  

P value for $B_2$ is less than 0.05 and hence H0 for $B_2$ can be rejected i.e. Y has a dependency on x2 as per the model. One more observation, even though $x_1$ and $x_2$ are corelated and should have had impact on the response (y) still the same is not visible in model summary, this suggests that the model is not performing properly due to multicolinearity.

## Now lets fit a least squares regression to predict $y$ using only $x_1$. Can we reject the null hypothesis $H_0$: $_1$ = 0?

```python
df_x_q7d = sm.add_constant(df_x1)

# Fit the regression model
model7d = sm.OLS(df_y, df_x_q7d).fit()

# Print the model summary
print(model7d.summary())
```

The estimated coefficients $B_0$ $\approx$ 2.24 and $B_1$ $\approx$ 1.87 are close to their true values, showing a good fit. Since the p-value is less than 0.05, we can reject the null hypothesis (H0) for $B_1$. This suggests that Y strongly depends on $X_1$ in the model.

## Lets fit a least squares regression to predict $y$ using only $x_2$. Can we reject the null hypothesis $H_0$: $_1$ = 0?

```python
df_x_q7e = sm.add_constant(df_x2)

# Fit the polynomial regression model
model7e = sm.OLS(df_y, df_x_q7e).fit()

# Print the model summary
print(model7e.summary())
```

The estimated coefficients $B_0$ $\approx$ 2.26 and $B_2$ $\approx$ 3.56 show that $B_0$ is close to the true intercept, but $B_2$ is quite different from its expected value. Since the p-value is less than 0.05, we can reject $H_0$ for $B_2$. This means $Y$ depends on $X_2$, but with a different coefficient than expected.


## Do the results obtained in (c)(e) contradict each other?

Yes, the results contradict each other because $X_1$ and $X_2$ are highly correlated, leading to collinearity. This makes it difficult to determine the independent effect of each predictor on $Y$. Collinearity increases the standard errors of $B_1$ and $B_2$, reducing the accuracy of their estimates. As a result, we fail to reject $H_0$ for $X_1$, and the true importance of $X_1$ and $X_2$ is masked.
