# Holy Quran Abjad Values Analysis

Many people have done some interesting analysis of the Holy Quran verses. I tried to take a different approach. Here is what I've done in the below code. The work is in progress, and we might not be able to reveal any significant patterns or trends. But I feel like this is a very interesting subject.

1. Every word in Arabic or any other language has a number value. In Arabic this is called Abjad Values
   1.1. For example Alif, Hamza, have abjad value 1
   1.2. Ya is 10, Kaf is 20, Qaf is 100, ghayn is 1000 and so on.
   1.3. You can think of it as a Numerical Encoding of Text Data.
2. For my analysis I encoded 2 of the Quranic Surah's (Fatiha and Ayat ul Kusi) in their Abjad Values and saved them as a Numpy Array in a txt file for each surah.
3. Than I wrote a program to load the specific Abjad value for the user selected surah.
4. After loading I plotted a Line Plot of all the abjad values to see if they reveal any patterns.


The work is still in progress. Besides the plots, I tried few other things as well but I'll not share those here for now until it gets matured.

## Import Packages

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import sys
import json

#import ipywidgets as widgets
from IPython.display import display
```

## Load the dictionary holding the key and file name for the currently encoded Abjad values chapters.

```python
with open('surah-names.json', 'r') as file:
    surah_file_names = json.load(file)

# Structure of the dictionay is, Key:[Surah Description, FileName]
list(surah_file_names.keys())
```

```python
selected_key = "kursi"
show_scatter_plot = True
show_line_plot = True
```

```python
input_file_name = surah_file_names.get(selected_key)[1]
surah_description = surah_file_names.get(selected_key)[0]
```

## Load the Abjad Values

```python
print(f'Selected Surah = {selected_key}')
print(f'Show ScatterPlot = {show_scatter_plot}')
print(f'Show Line Plot = {show_line_plot}')

surah_abjad_value = np.loadtxt(input_file_name)
```

## Print the Abjad Values

```python
surah_abjad_value[0:10]
```

## Draw the scatter and/or line plot

```python
# Create a scatter plot for character-by-character Abjad values
plt.figure(figsize=(10, 6))
#plt.scatter(range(len(abjad_values_fatiha_char)), abjad_values_fatiha_char, color='blue', marker='o')

if show_scatter_plot:
    plt.scatter(range(len(surah_abjad_value)), surah_abjad_value, color='blue', marker='o', s=5)

if show_line_plot:
    plt.plot(surah_abjad_value, linestyle='--', color='green')

if show_scatter_plot and show_line_plot:
    plt.twinx()
    
# Add labels and title
plt.title(f'Character-by-Character Abjad Values of {surah_description}')
plt.xlabel('Character Index')
plt.ylabel('Abjad Value')

# Display the plot
plt.show()
```
